**Ejercicio 1**

Modifica el programa que resuelve la ecuación cuadratica (vista en la ayudantía), de tal forma que tenga solución cuando el discriminante sea menor a cero.

```python
import math

def cuadrado(x):
    y = x**2
    return y

def ecuacion_cuadratica(a, b, c):

    d = cuadrado(b) - 4*a*c

    if d >= 0:
        raiz = math.sqrt(d)
    else:
        raiz = (-d)**0.5 * 1j   # raíz imaginaria

    x1 = (-b + raiz) / (2*a)
    x2 = (-b - raiz) / (2*a)

    return x1, x2


a = float(input("Introduce a: "))
b = float(input("Introduce b: "))
c = float(input("Introduce c: "))

sol1, sol2 = ecuacion_cuadratica(a, b, c)

print("Las soluciones son:")
print(sol1)
print(sol2)
```

**Ejercicio 2**

Desarrollar un programa que calcule la suma de los dígitos de cualquier número entero positivo introducido por el usuario.*italicized text*

```python
numero = int(input("Introduce un número entero positivo: "))

suma = 0

while numero > 0:
    digito = numero % 10      # obtiene el último dígito
    suma = suma + digito      # suma el dígito
    numero = numero // 10     # elimina el último dígito

print("La suma de los dígitos es:", suma)
```

**Ejercicio 3**

Escribir un programa que pida al usuario una palabra y muestre por pantalla si es un palíndromo.


```python
palabra = input("Introduce una palabra: ")

inicio = 0
fin = len(palabra) - 1
es_palindromo = True

while inicio < fin:
    if palabra[inicio] != palabra[fin]:
        es_palindromo = False
        break

    inicio = inicio + 1
    fin = fin - 1

if es_palindromo:
    print("La palabra es un palíndromo")
else:
    print("La palabra no es un palíndromo")
```
